### 2.Import pandas and read Salaries.csv file

```python
import pandas as pd
df = pd.read_csv('Salaries.csv')
df.drop(['Status','Notes'], axis = 1, inplace = True)
df['BasePay'].fillna(1000, inplace = True)
df['Benefits'].fillna(1000, inplace = True)
```

### 11. Display relationship between OvertimePay and BasePay
* Use scatter in matplotlib

```python
import matplotlib.pyplot as plt
plt.scatter(df['BasePay'], df['OvertimePay'])
plt.title('Relationship between OvertimePay and BasePay')
plt.xlabel('BasePay')
plt.ylabel('OvertimePay')
plt.show()
```
