## Control line styles

All the available line styles are available in the following table:


**Style abbreviation**  **Style**

-                    solid line
   
--                   dashed line
   
-.                   dash-dot line
   
:                    dotted line
   


Now, we can see the default format string for a single line plot is 'b-'.

```python
x16 = np.arange(1, 5)
# plot blue line with -- , red line with -. , green line with :

plt.plot(x16, lw=1, ls='--', color = 'green')
plt.plot(x16+1, lw=1, ls='-', color = 'red')
plt.plot(x16+2, lw=1, ls=':', color = 'blue')
```

```python
x = np.linspace(0, 2)
# plot x and x and label it as 'linear'
# plot x and x**2 and label it as 'quadratic'
# plot x and x**3 and label it as cubic
# label x and y axes
# show legend

```

```python
plt.plot(x,x, lw=1, ls='--', color = 'blue', label = 'linear')
plt.plot(x,x**2, lw=1, ls='-', color = 'red', label = 'quadratic' )
plt.plot(x,x**3, lw=1, ls=':', color = 'green', label = 'cubic')
plt.legend()
```

### 3. Find correlation between variables. Then use heatmap for visualization (annot=True - numbers will visually appear in heatmap)

```python
dataset.corr()
```

```python
sns.heatmap(dataset.corr(), annot = True)
```

### 4. Display relationship between highway miles per gallon and city miles per gallon
* Use scatter in matplotlib
* Interpret how variables correlated according to scatter

```python
plt.scatter(dataset['hwy'], dataset['cty'])
plt.title('Relationship between highway miles per gallon and city miles per gallon')
plt.xlabel('Highway miles per gallon')
plt.ylabel('City miles per gallon')
plt.show()
```

### 5. Show distributions and scatters between all variables by type of drive train
*  Which variables have the highest correlation? Guess it visually.
*  Which variables have the lowest correlation? Guess it visually.

```python
sns.pairplot(dataset)
# the highest correlation between variables : 
# hwy and cty
# the lowest crrelation between variables :
# hwy and disply
# cty and disply
```

### 6. Which class has most observed in dataset. 
* Use countplot

```python
sns.countplot(x='class',data=dataset)
```

### 7. Visualize engine displacement by each class 
* Use boxplot

```python
sns.boxplot(x="class", y="displ", data=dataset ,palette='rainbow')
```

### 8. Do the same thing in violinplot  

```python
sns.violinplot(x="class", y="displ", data=dataset,palette='rainbow')
```
